```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Loading & Initial Inspection

```python
# Load data
df = pd.read_csv('drive/MyDrive/games.csv')
```

```python
# Initial inspection
print(f"Shape: {df.shape}")
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Distribution')
plt.show()
```

# Data Cleaning & Transformation

```python
# Convert 'Price' column to float after removing dollar sign
df['Price'] = df['Price'].replace('[\$,]', '', regex=True).astype(float)

# Convert 'Size' to megabytes (MB)
df['Size'] = pd.to_numeric(df['Size'], errors='coerce') / 1e6

# Parse release date columns with dayfirst format
df['Original Release Date'] = pd.to_datetime(df['Original Release Date'], dayfirst=True, errors='coerce')
df['Current Version Release Date'] = pd.to_datetime(df['Current Version Release Date'], dayfirst=True, errors='coerce')

# Extract year from parsed dates
df['Original Release Year'] = df['Original Release Date'].dt.year
df['Current Version Year'] = df['Current Version Release Date'].dt.year

# Handle missing values
df['Subtitle'] = df['Subtitle'].fillna('Not Available')
df['In-app Purchases'] = df['In-app Purchases'].fillna('No')
```

# Univariate Analysis

```python
plt.figure(figsize=(10,6))
sns.histplot(df[df['Price'] > 0]['Price'], bins=20, kde=True)
plt.title('Distribution of Paid Game Prices')
plt.xlabel('Price ($)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Average User Rating'].dropna(), bins=10)
plt.title('User Rating Distribution')
plt.xlabel('Rating (1-5)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Size'].dropna())
plt.title('Game Size Distribution (MB)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['Age Rating'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Age Rating Distribution')
plt.ylabel('')
plt.show()
```

# Bivariate Analysis

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Price', y='Average User Rating', data=df)
plt.title('Price vs. User Rating')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Original Release Year', y='Average User Rating',
             data=df[df['Original Release Year'] > 2000])
plt.title('Ratings by Release Year')
plt.show()
```

```python
# Extract primary genres
genre_counts = df['Primary Genre'].value_counts().head(5)

plt.figure(figsize=(12,6))
genre_counts.plot(kind='bar')
plt.title('Top 5 Game Genres')
plt.ylabel('Count')
plt.show()
```

# Multivariate Analysis

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='Size', y='Price', hue='Average User Rating',
                size='User Rating Count', data=df)
plt.title('Game Size vs Price Colored by Rating')
plt.xlabel('Size (MB)')
plt.show()
```

```python
iap_analysis = df.groupby('In-app Purchases').agg(
    Avg_Rating=('Average User Rating', 'mean'),
    Avg_Price=('Price', 'mean'),
    Count=('ID', 'count')
).reset_index()

print(iap_analysis)
```
